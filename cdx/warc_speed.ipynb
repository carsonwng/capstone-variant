```python
import requests
import time
import gzip
```

```python
url = "https://data.commoncrawl.org/cc-index/collections/CC-MAIN-2023-23/indexes/cdx-00001.gz"
```

```python
# approach 1

s_time = time.time()

with requests.Session() as session:
    with session.get(url, stream=True) as r:
        r.raise_for_status()

        with open("tmp.gz", "wb") as f:
            f.write(r.content)

        left = 10
        with gzip.open("tmp.gz", "rt") as f:
            for line in f:
                if left > 5:
                    left -= 1
                    print(line)

print("approach 1: ", time.time() - s_time)
```

```python
# making the request portion asynchronous and running two jobs

import asyncio
import aiohttp

s_time = time.time()

urls = [
    "https://data.commoncrawl.org/cc-index/collections/CC-MAIN-2023-23/indexes/cdx-00001.gz",
    "https://data.commoncrawl.org/cc-index/collections/CC-MAIN-2023-23/indexes/cdx-00002.gz"
]

async def get(t_url: str):
    session = aiohttp.ClientSession()

    res = await session.get(t_url, raise_for_status=True)
    
    with open(f"{t_url.split('/')[-1]}", "wb") as f:
        f.write(await res.content.read())
    
    session.close()

await asyncio.gather(
    get(urls[0]),
    get(urls[1])
)

print("approach 2 (2 urls): ", time.time() - s_time)
```
